### Linear regression 1 variable

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('ex1data1.txt')
data.head()
```

```python
x_train = data.values[:, 0]
y_train = data.values[:, 1]
m = len(x_train)
```

```python
plt.scatter(x_train, y_train, marker='.')
```

```python
# TODO: hypothesis function
def hypothesis(x, theta0, theta1):
    return theta0 + x * theta1

# TODO: cost function
def cost_function(x, y, theta0, theta1):
    sum_error = 0
    for i in range(m):
        sum_error += pow(((theta0 + theta1 * x[i]) - y[i]), 2)
    
    return (sum_error / (2 * m))

# TODO: gradient descent
def gradient_descent(x, y, theta0, theta1, alpha):
    theta0_gradient = 0
    theta1_gradient = 0

    for i in range(m):
        theta0_gradient += ((theta0 + theta1 * x[i]) - y[i])
        theta1_gradient += (x[i] * ((theta0 + theta1 * x[i]) - y[i]))
    
    theta0 -= alpha * (theta0_gradient / m)
    theta1 -= alpha * (theta1_gradient / m)

    return theta0, theta1

def train(x, y, alpha, theta0, theta1, iter):
    J_his = []
    for i in range(iter):
        theta0, theta1 = gradient_descent(x, y, theta0, theta1, alpha)
        J_his.append(cost_function(x, y, theta0, theta1))

    return theta0, theta1, J_his
```

```python
theta0, theta1, cost = train(x_train, y_train, 0.01, 0, 0, 1500)
print(theta0)
print(theta1)
#print(cost)
```
